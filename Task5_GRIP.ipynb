```python
#Most preferred ship mode
plt.figure(figsize=(8,5))
sns.countplot(data['Ship_Mode'])
```

```python
plt.figure(figsize=(8,5))
sns.set(style="whitegrid")
sns.barplot(x='Region',y='Profit',data=data)
plt.title("Region wise Profit")
plt.show()
```

```python
data['City'].value_counts().head(50)
```

```python
data['City'].value_counts().head(50).plot(kind='bar',figsize=(20,8),color='red')
plt.ylabel('Number of sales')
plt.xlabel('City')
plt.title("City wise Sales")
plt.show()
```

```python
sns.countplot(data['State']).set_xticklabels(sns.countplot(data['State']).get_xticklabels(), rotation=90, ha="right")
plt.ylabel('Number of sales')
plt.xlabel('states')
fig = plt.gcf()
fig.set_size_inches(20,8)
plt.title("State wise Sales")
plt.show()
```

```python
#Most profitable category and sub-catergory
cat_sub=pd.DataFrame(data.groupby(['Category','Sub_Category']).sum()['Profit']) #grouping categry and subcategory
```

```python
cat_sub.sort_values(['Category','Sub_Category'],ascending=False) #sort the result
```
